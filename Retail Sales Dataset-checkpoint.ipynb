```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, mean_absolute_percentage_error

# For simple models
from sklearn.linear_model import LinearRegression
from xgboost import XGBRegressor

# ────────────────────────────────────────────────
# 1. Load & prepare monthly data
# ────────────────────────────────────────────────
df = pd.read_csv('train.csv')

# Fix date format (your data uses dd-mm-yy)
df['Order Date'] = pd.to_datetime(df['Order Date'], format='%d-%m-%y', errors='coerce')
df = df.dropna(subset=['Order Date'])  # drop any bad dates

# Aggregate to monthly total sales
monthly = (
    df.groupby(pd.Grouper(key='Order Date', freq='ME'))['Sales']  # ME = month end
    .sum()
    .reset_index()
)

monthly = monthly.rename(columns={'Order Date': 'ds', 'Sales': 'y'})
monthly['ds'] = pd.to_datetime(monthly['ds'])
monthly = monthly.set_index('ds').sort_index()

print("Monthly data shape:", monthly.shape)
print(monthly.head(8))

# Quick plot
plt.figure(figsize=(12, 5))
monthly['y'].plot(title='Monthly Total Sales – Raw Data', marker='o')
plt.grid(alpha=0.3)
plt.ylabel('Sales ($)')
plt.show()




# ────────────────────────────────────────────────
# 2. Feature engineering
# ────────────────────────────────────────────────
df_month = monthly.copy()

# Lags (previous months)
for lag in [1, 2, 3, 6, 12]:
    df_month[f'lag_{lag}'] = df_month['y'].shift(lag)

# Rolling statistics
df_month['rolling_mean_3'] = df_month['y'].rolling(3).mean()
df_month['rolling_mean_6'] = df_month['y'].rolling(6).mean()

# Calendar features
df_month['month']     = df_month.index.month
df_month['year']      = df_month.index.year
df_month['quarter']   = df_month.index.quarter
df_month['is_q4']     = (df_month['quarter'] == 4).astype(int)  # holiday season effect

# Trend feature (simple linear trend)
df_month['trend'] = np.arange(len(df_month))

df_month = df_month.dropna().copy()

print("Features created. Shape after dropna:", df_month.shape)





# ────────────────────────────────────────────────
# 3. Train-test split (last 12 months = test)
# ────────────────────────────────────────────────
test_size = 12
train = df_month.iloc[:-test_size]
test  = df_month.iloc[-test_size:]

features = [col for col in df_month.columns if col != 'y']

X_train, y_train = train[features], train['y']
X_test,  y_test  = test[features],  test['y']

# Baseline: repeat last year's value (seasonal naive)
naive_pred = X_test['lag_12']

print("Naive Seasonal MAE  :", mean_absolute_error(y_test, naive_pred))
print("Naive Seasonal MAPE :", mean_absolute_percentage_error(y_test, naive_pred)*100, "%")




# ────────────────────────────────────────────────
# 4. Models
# ────────────────────────────────────────────────
models = {
    'Linear Regression': LinearRegression(),
    'XGBoost': XGBRegressor(n_estimators=300, learning_rate=0.07, max_depth=4, random_state=42)
}

results = []

for name, model in models.items():
    model.fit(X_train, y_train)
    pred = model.predict(X_test)
    
    mae  = mean_absolute_error(y_test, pred)
    mape = mean_absolute_percentage_error(y_test, pred) * 100
    rmse = np.sqrt(mean_squared_error(y_test, pred))
    
    results.append({
        'Model': name,
        'MAE': mae,
        'MAPE (%)': mape,
        'RMSE': rmse
    })
    
    # Plot prediction
    plt.figure(figsize=(12,5))
    plt.plot(test.index, y_test, label='Actual', marker='o', linewidth=2)
    plt.plot(test.index, pred, label=f'{name} pred', marker='x', linestyle='--')
    plt.plot(test.index, naive_pred, label='Seasonal Naive', linestyle=':', color='gray')
    plt.title(f'{name} – Actual vs Predicted Monthly Sales')
    plt.legend()
    plt.grid(alpha=0.3)
    plt.ylabel('Sales ($)')
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()

# Summary table
print(pd.DataFrame(results).round(0).to_string(index=False))
```

```python
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.preprocessing import LabelEncoder
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
from xgboost import XGBClassifier



# ────────────────────────────────────────────────
# 2. Create customer-level features
# ────────────────────────────────────────────────
customer_agg = df.groupby('Customer ID').agg(
    total_orders     = ('Order ID', 'nunique'),
    total_spend      = ('Sales', 'sum'),
    first_order      = ('Order Date', 'min'),
    last_order       = ('Order Date', 'max'),
    distinct_categories = ('Category', 'nunique'),
    region_mode      = ('Region', lambda x: x.mode()[0] if not x.mode().empty else np.nan),
    segment          = ('Segment', 'first'),           # usually stable per customer
    state_mode       = ('State', lambda x: x.mode()[0] if not x.mode().empty else np.nan),
).reset_index()

# Time-based features
customer_agg['tenure_days'] = (customer_agg['last_order'] - customer_agg['first_order']).dt.days
customer_agg['recency_days'] = (pd.Timestamp.now() - customer_agg['last_order']).dt.days

# ────────────────────────────────────────────────
# 3. Category preference (what we want to predict)
# ────────────────────────────────────────────────
cat_pivot = df.pivot_table(
    index='Customer ID',
    columns='Category',
    values='Sales',
    aggfunc='sum',
    fill_value=0
).reset_index()

cat_pivot.columns = ['Customer ID', 'Furniture_sales', 'Office Supplies_sales', 'Technology_sales']

# Normalize → % of spend per category
cat_pivot['total_cat_sales'] = cat_pivot.iloc[:,1:4].sum(axis=1)
for col in ['Furniture_sales', 'Office Supplies_sales', 'Technology_sales']:
    cat_pivot[col + '_pct'] = cat_pivot[col] / cat_pivot['total_cat_sales'].replace(0, np.nan)

# Target = category with highest $ spend
cat_pivot['preferred_category'] = cat_pivot[['Furniture_sales', 'Office Supplies_sales', 'Technology_sales']].idxmax(axis=1)
cat_pivot['preferred_category'] = cat_pivot['preferred_category'].str.replace('_sales', '')

# Merge back
customer_features = customer_agg.merge(cat_pivot[['Customer ID', 'preferred_category', 
                                                  'Furniture_sales_pct', 'Office Supplies_sales_pct', 'Technology_sales_pct']],
                                       on='Customer ID', how='left')

# ────────────────────────────────────────────────
# 4. Final modeling dataset
# ────────────────────────────────────────────────
features = [
    'total_orders', 'total_spend', 'tenure_days', 'recency_days',
    'distinct_categories', 'Furniture_sales_pct', 'Office Supplies_sales_pct', 'Technology_sales_pct'
]

X = customer_features[features].copy()
y = customer_features['preferred_category'].copy()

# Encode target
le = LabelEncoder()
y_encoded = le.fit_transform(y)

X_train, X_test, y_train, y_test = train_test_split(X, y_encoded, test_size=0.25, random_state=42, stratify=y_encoded)

print("Class distribution:\n", pd.Series(le.inverse_transform(y_encoded)).value_counts(normalize=True))

# ────────────────────────────────────────────────
# 5. Train & evaluate models
# ────────────────────────────────────────────────
models = {
    'Logistic': LogisticRegression(max_iter=1000, multi_class='multinomial'),
    'Random Forest': RandomForestClassifier(n_estimators=200, random_state=42, class_weight='balanced'),
    'XGBoost': XGBClassifier(use_label_encoder=False, eval_metric='mlogloss', random_state=42)
}

for name, model in models.items():
    model.fit(X_train, y_train)
    preds = model.predict(X_test)
    acc = accuracy_score(y_test, preds)
    print(f"\n{name} Accuracy: {acc:.3f}")
    print(classification_report(y_test, preds, target_names=le.classes_))

# ────────────────────────────────────────────────
# 6. Confusion matrix (visual)
# ────────────────────────────────────────────────
best_model = models['XGBoost']  # or whichever performed best
cm = confusion_matrix(y_test, best_model.predict(X_test))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
            xticklabels=le.classes_, yticklabels=le.classes_)
plt.title('Confusion Matrix - Category Preference Prediction')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import mean_absolute_error, mean_absolute_percentage_error
from xgboost import XGBRegressor

# Load & clean date
df = pd.read_csv('train.csv')
df['Order Date'] = pd.to_datetime(df['Order Date'], format='%d-%m-%y', errors='coerce')
df = df.dropna(subset=['Order Date'])

print("Date range:", df['Order Date'].min(), "–", df['Order Date'].max())


```

```python
def forecast_group(group_col, group_values, horizon=12):
    results = {}
    
    for val in group_values:
        subset = df[df[group_col] == val].copy()
        monthly = subset.groupby(pd.Grouper(key='Order Date', freq='ME'))['Sales'].sum().reset_index()
        monthly = monthly.rename(columns={'Order Date': 'ds', 'Sales': 'y'})
        monthly['ds'] = pd.to_datetime(monthly['ds'])
        monthly = monthly.set_index('ds').sort_index()
        
        if len(monthly) < 24: 
            print(f"Skipping {val} — too few months ({len(monthly)})")
            continue
        
        # Features
        df_m = monthly.copy()
        for lag in [1, 2, 3, 6, 12]:
            df_m[f'lag_{lag}'] = df_m['y'].shift(lag)
        df_m['rolling_mean_3'] = df_m['y'].rolling(3).mean()
        df_m['month'] = df_m.index.month
        df_m['year'] = df_m.index.year
        df_m['is_q4'] = (df_m.index.quarter == 4).astype(int)
        df_m['trend'] = np.arange(len(df_m))
        df_m = df_m.dropna()
        
        # Split
        test_size = min(horizon, len(df_m)//4)
        train = df_m.iloc[:-test_size]
        test  = df_m.iloc[-test_size:]
        
        features = [c for c in df_m.columns if c != 'y']
        X_train, y_train = train[features], train['y']
        X_test, y_test   = test[features], test['y']
        
        # Model
        model = XGBRegressor(n_estimators=200, learning_rate=0.08, max_depth=4, random_state=42)
        model.fit(X_train, y_train)
        pred = model.predict(X_test)
        
        mae = mean_absolute_error(y_test, pred)
        mape = mean_absolute_percentage_error(y_test, pred) * 100
        
        results[val] = {
            'mae': mae,
            'mape': mape,
            'test_actual': y_test,
            'test_pred': pred,
            'test_index': test.index
        }
        
        print(f"{val:8} → Test MAE: ${mae:,.0f} | MAPE: {mape:.1f}%")
    
    return results


# Run for regions
region_forecasts = forecast_group('Region', ['West', 'East', 'Central', 'South'], horizon=12)


# ─── Plot all regions ────────────────────────────────────────
plt.figure(figsize=(14, 8))

for region, info in region_forecasts.items():
    plt.plot(info['test_index'], info['test_actual'], label=f'{region} Actual', marker='o', linewidth=2, alpha=0.8)
    plt.plot(info['test_index'], info['test_pred'],  label=f'{region} Pred',  linestyle='--', marker='x')

plt.title('Monthly Sales Forecast by Region – Last 12 months', fontsize=14, pad=15)
plt.ylabel('Sales ($)')
plt.grid(alpha=0.3)
plt.legend(ncol=2)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python

from prophet import Prophet
from sklearn.metrics import mean_absolute_error, mean_absolute_percentage_error

# ────────────────────────────────────────────────
# Helper: prepare prophet-style dataframe for any group
# ────────────────────────────────────────────────
def prepare_prophet_df(subset_df):
    monthly = (
        subset_df.groupby(pd.Grouper(key='Order Date', freq='ME'))['Sales']
        .sum()
        .reset_index(name='y')
    )
    monthly = monthly.rename(columns={'Order Date': 'ds'})
    monthly['ds'] = pd.to_datetime(monthly['ds'])
    monthly = monthly.sort_values('ds')
    return monthly


# ────────────────────────────────────────────────
# Function: fit Prophet + forecast + baseline
# ────────────────────────────────────────────────
def forecast_series(series_df, name="Series", periods=12, changepoint_prior_scale=0.05):
    if len(series_df) < 24:
        print(f"{name}: too few data points ({len(series_df)} months)")
        return None, None, None

    # Train / last 12 months test split
    train = series_df.iloc[:-12].copy()
    test  = series_df.iloc[-12:].copy()

    # ─── Prophet model ────────────────────────────────────────
    m = Prophet(
        yearly_seasonality=True,
        weekly_seasonality=False,
        daily_seasonality=False,
        changepoint_prior_scale=changepoint_prior_scale,
        seasonality_mode='multiplicative'   # better for sales with growing trend
    )

    # Optional: add Q4 effect as custom seasonality / regressor
    train['is_q4'] = (train['ds'].dt.quarter == 4).astype(int)
    test['is_q4']  = (test['ds'].dt.quarter == 4).astype(int)
    m.add_regressor('is_q4')

    m.fit(train)

    # Future dataframe (includes test period + next 12 months)
    future = m.make_future_dataframe(periods=periods, freq='ME')
    future['is_q4'] = (future['ds'].dt.quarter == 4).astype(int)

    forecast = m.predict(future)

    # Extract test period predictions
    forecast_test = forecast[forecast['ds'].isin(test['ds'])]
    pred_test = forecast_test['yhat'].values

    # Metrics on test set
    mae  = mean_absolute_error(test['y'], pred_test)
    mape = mean_absolute_percentage_error(test['y'], pred_test) * 100

    print(f"{name:12}  Test MAE: ${mae:,.0f}   MAPE: {mape:.1f}%")

    return forecast, test, pred_test


# ────────────────────────────────────────────────
# 1. Total sales forecast
# ────────────────────────────────────────────────
total_df = prepare_prophet_df(df)
total_forecast, total_test, total_pred_test = forecast_series(
    total_df, name="Total", periods=12
)

# ────────────────────────────────────────────────
# 2. Region forecasts
# ────────────────────────────────────────────────
regions = ['West', 'East', 'Central', 'South']
region_forecasts = {}
region_test_data = {}

for r in regions:
    subset = df[df['Region'] == r]
    reg_df = prepare_prophet_df(subset)
    fc, tst, pred = forecast_series(reg_df, name=r, periods=12)
    if fc is not None:
        region_forecasts[r] = fc
        region_test_data[r] = tst


# ────────────────────────────────────────────────
# 3. Simple baseline: average of same month in previous years
# ────────────────────────────────────────────────
def same_month_average_baseline(full_df):
    # Assuming 'ds' is your date column from the Prophet DataFrame
    full_df['month'] = full_df['ds'].dt.month
    monthly_avg = full_df.groupby('month')['y'].mean()
    
    # Get the last 12 months and map their month numbers to the average
    test_months = full_df.iloc[-12:]['ds'].dt.month
    baseline_pred = test_months.map(monthly_avg)
    
    return baseline_pred.values


total_baseline = same_month_average_baseline(total_df)

print("\nBaseline (same-month avg previous years) MAPE:",
      mean_absolute_percentage_error(total_test['y'], total_baseline)*100, "%")
```
